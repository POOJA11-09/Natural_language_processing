```python
import re

text = """
Natural Language Processing, also known as NLP, is one of the most important areas of Artificial Intelligence.
Today, companies use NLP in chatbots, search engines, recommendation systems, social media analytics,
customer feedback analysis, automatic translation, and voice assistants.

In 2026, many organizations are collecting huge amounts of text data from websites, emails, reviews,
online forms, tweets, product comments, and support tickets. However, this text data is often messy.
For example, a customer may write: "I LOVE this product!!! Visit https://example.com/product123 #Amazing #AI".
Another user may write: "The service was not good... I waited 45 minutes and no one replied."

Before applying machine learning or deep learning models, we need to clean and prepare the text.
This step is called text preprocessing. It may include converting text to lowercase, removing URLs,
removing punctuation, removing numbers, tokenizing sentences into words, removing stop words,
stemming, lemmatization, and sometimes removing emojis or special characters.

Good preprocessing can improve the quality of NLP models. However, preprocessing should be done carefully.
For example, removing the word "not" from the sentence "I am not happy" can change the meaning completely.
Similarly, removing hashtags may not always be a good idea in social media analytics because hashtags
can contain useful information about topics, opinions, or trends.
"""

text_clean=re.sub(r'http\S+', '', text)

text_clean=re.sub(r'!+', '', text_clean)

text_clean=re.sub(r'#(\w+)', r'\1', text_clean)

text_clean=re.sub(r'\s+', ' ', text_clean).strip()

print("STEP 1 OUTPUT:\n", text_clean)
```

```python
text_lower=text_clean.lower()
print("\nSTEP 2 OUTPUT:\n", text_lower)
```

```python
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize

tokens=word_tokenize(text_lower)
print("\nSTEP 3 OUTPUT (TOKENS):\n", tokens)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words=set(stopwords.words('english'))
stop_words.remove('no')
stop_words.remove('not')

filtered_tokens=[w for w in tokens if w.isalpha() and w not in stop_words]
print("\nSTEP 4 OUTPUT (STOPWORDS REMOVED):\n", filtered_tokens)
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
nltk_lemmas = [lemmatizer.lemmatize(w) for w in filtered_tokens]

print("\nSTEP 5 OUTPUT (NLTK LEMMAS):\n", nltk_lemmas)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
spacy_lemmas = [nlp(token)[0].lemma_ for token in filtered_tokens]

print("\nSTEP 6 OUTPUT (spaCy LEMMAS):\n", spacy_lemmas)
```

```python
import pandas as pd

comparison=[]

unique_map={}

for original, n_lemma, s_lemma in zip(filtered_tokens, nltk_lemmas, spacy_lemmas):
    if original not in unique_map:
        unique_map[original]=(n_lemma, s_lemma)

# Now compare only unique words
for original, (n_lemma, s_lemma) in unique_map.items():
    if n_lemma != s_lemma:
        comparison.append([original, n_lemma, s_lemma])

df=pd.DataFrame(comparison, columns=["Original Token", "NLTK Lemma", "spaCy Lemma"])

print("\nSTEP 7 OUTPUT (UNIQUE WORD COMPARISON):")
print(df)
```
