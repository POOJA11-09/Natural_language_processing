```python
import spacy
from spacy import displacy
from IPython.display import HTML, display

#   Load the English language model
nlp = spacy.load("en_core_web_sm")

text = "I saw the man with a telescope."

doc = nlp(text)

# Print the dependency parse information for each token
for token in doc:
    print(f"{token.text:<12} {token.dep_:<12} {token.head.text}")

html = displacy.render(doc, style="dep", jupyter=False)

# Display the dependency parse visualization in Jupyter Notebook
display(HTML(html))

# Discuss with your partner:
# 1. What is the ROOT?
# 2. What does "with a telescope" modify?
# 3. Does the sentence have more than one meaning?
# 4. Did spaCy choose the interpretation you expected?
# 5. Rewrite the sentence to remove ambiguity.
```

1.The ROOT word is "saw" because it is the main verb of the sentence.

2.The phrase "with a telescope" modifies "saw" in spaCy’s parse means the telescope is being used to see the man.

3.Yes, the sentence has more than one meaning. It is ambiguous because:
   The speaker used a telescope to see the man.
   The man had a telescope.

4.SpaCy chose the interpretation  "speaker used the telescope to see the man". This may or may not match the expected interpretation.

5.The sentence can be rewritten to remove ambiguity
I used a telescope to see the man.
I saw the man who was carrying a telescope.

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """
Microsoft announced a new AI research initiative
in Berlin on 15 June 2026.

The project involves collaboration with Google,
SAP and SRH University Heidelberg.

Satya Nadella will present the initiative
at a conference in Germany.
"""

doc = nlp(text)

for ent in doc.ents:
    print(
        ent.text,
        ent.label_,
        spacy.explain(ent.label_)
    )

#1. Which entities were detected?
#2. Which labels were assigned?
#3. Which entity types occur most often?
#4. Were any entities missed?
```

1. The entities detected are:

   * Microsoft
   * AI
   * Berlin
   * 15 June 2026
   * Google
   * SAP
   * SRH University Heidelberg
   * Satya Nadella
   * Germany

2. The labels assigned are:

   * Microsoft :ORG
   * AI :ORG
   * Berlin :GPE
   * 15 June :2026 DATE
   * Google :ORG
   * SAP :ORG
   * SRH University Heidelberg :ORG
   * Satya Nadella :PERSON
   * Germany :GPE

3. The entity type that occurs most often is ORG because several company and institution names were detected.

4. Some entities may have been missed or labeled differently depending on the spaCy model. For example:

   * “AI research initiative” may not be recognized as a complete entity and may noy recieve the expected labelto the word "AI".
   * “SRH University Heidelberg” could be partially detected.


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [
    "AI is transforming healthcare",
    "Machine learning improves healthcare",
    "Football is a popular sport"
]

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)

print("Vocabulary:")
print(vectorizer.get_feature_names_out())

print("\nTF-IDF Matrix:")
print(tfidf_matrix.toarray())

similarity_matrix = cosine_similarity(tfidf_matrix)

print("\nCosine Similarity Matrix:")
print(similarity_matrix)

print("Doc1 vs Doc2:", round(similarity_matrix[0][1], 3))
print("Doc1 vs Doc3:", round(similarity_matrix[0][2], 3))
print("Doc2 vs Doc3:", round(similarity_matrix[1][2], 3))

# 1. Which pair has the highest similarity?
# 2. Is this result reasonable?
# 3. Why does TF-IDF treat some words as more important?
# 4. Does cosine understand meaning?
# 5. Would it know that "AI" and "machine learning" are related?
```

1.The pair with the highest similarity is Document 1 and Document 2 because both discuss healthcare and related technology topics.

2.Yes, this result is reasonable because the two documents share similar words such as “healthcare” and are both related to artificial intelligence and technology. Document 3 is about football, so it is less similar.

3.TF-IDF treats some words as more important because it gives higher weight to words that are frequent in one document but rare across all documents. Common words that appear everywhere get lower importance.

4.No, cosine similarity does not truly understand meaning it only measures similarity based on word frequency and shared terms in the documents.

5.No, basic TF-IDF cannot understand that “AI” and “machine learning” are closely related concepts unless they contain similar words. It only compares documents based on the words they share, not on the actual meaning of the text.
