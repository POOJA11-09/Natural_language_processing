```python
import spacy
from spacy import displacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")

text = "I saw the man with a telescope."

doc = nlp(text)

# Print the dependency parse information for each token
for token in doc:
    print(f"{token.text:<12} {token.dep_:<12} {token.head.text}")

html = displacy.render(doc, style="dep", jupyter=False)

# Display the dependency parse visualization in Jupyter Notebook
display(html)


```

1. What is the ROOT?

    * "saw" as it is the main verb




2. What does "with a telescope" modify?

     * "with a telescope" modifies "saw" as it indicates how the action of seeing was performed.



3. Does the sentence have more than one meaning?

     *  Yes, the sentence can be interpreted in two ways:
     a) The man has a telescope and I saw him using it.
     b) I used a telescope to see the man.



4. Did spaCy choose the interpretation you expected?

     *  It does not choose one interpretation over another as "with a telescope" modifies "saw".



5. Rewrite the sentence to remove ambiguity.

     *  I saw the man who was using a telescope.

     *  Using a telescope, I saw the man.

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """
Microsoft announced a new AI research initiative
in Berlin on 15 June 2026.

The project involves collaboration with Google,
SAP and SRH University Heidelberg.

Satya Nadella will present the initiative
at a conference in Germany.
"""

doc = nlp(text)

for ent in doc.ents:
    print(
        ent.text,
        ent.label_,
        spacy.explain(ent.label_)
    )

```

1. Which entities were detected?

     *  "Microsoft", "Berlin", "15 June 2026", "Google", "SAP", "SRH University Heidelberg", "Satya Nadella", "Germany"

2. Which labels were assigned?

     *  "ORG" for organizations (Microsoft, Google, SAP, SRH University Heidelberg), 
        "GPE" for geopolitical entities (Berlin, Germany), 
        "DATE" for dates (15 June 2026),  
        "PERSON" for people (Satya Nadella).

3. Which entity types occur most often?

     *  "ORG"

4. Were any entities missed?

     *  "AI research initiative" and "conference in Germany" entities were missed depending upon the context

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [
    "AI is transforming healthcare",
    "Machine learning improves healthcare",
    "Football is a popular sport"
]

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)

print("Vocabulary:")
print(vectorizer.get_feature_names_out())

print("\nTF-IDF Matrix:")
print(tfidf_matrix.toarray())

similarity_matrix = cosine_similarity(tfidf_matrix)

print("\nCosine Similarity Matrix:")
print(similarity_matrix)

print("Doc1 vs Doc2:", round(similarity_matrix[0][1], 3))
print("Doc1 vs Doc3:", round(similarity_matrix[0][2], 3))
print("Doc2 vs Doc3:", round(similarity_matrix[1][2], 3))

```

1. Which pair has the highest similarity?

     * Doc1 vs Doc2 has the highest similarity with a score of 0.707

2. Is this result reasonable?

     * Yes , as Doc! and Doc2 both dicuss healthcare and the impact of AI and machine learning on it while Doc3 is about AI

3. Why does TF-IDF treat some words as more important?

     *  TF-IDF treats some words as more important based on their frequency in a document relative to their frequency across all documents. Words that appear frequently in a specific document but are rare across the entire corpus are given higher importance, while common words that appear in many documents are down-weighted. This helps to highlight the unique terms that are more relevant to the content of each document.

4. Does cosine understand meaning?

     *  No, it only measures the angle between two vectors in a high-dimensional space. It captures the similarity of the term usage and frequency but does not comprehend the context of the words.

5. Would it know that "AI" and "machine learning" are related?

     *  No, it would not know that they are related unless they occur in the same documents.
